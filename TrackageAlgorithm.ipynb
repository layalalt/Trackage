```python
class Bag:  
    
    def assign_to_reservation(self, reservation):
        self.reservation = reservation
        reservation.add_bag(self)   
        
    def __init__(self, bag_id, reservation, path):
        self.bag_id = bag_id
        self.ready = False
        self.assign_to_reservation(reservation)
        self.path = path

    def __str__(self):
        reservation_id = self.reservation.reservation_id if self.reservation else "None"
        return f"Bag(ID: {self.bag_id}, Reservation: {reservation_id})"

    def set_ready(self):
        self.ready = True   
```

```python
class Reservation:
    
    def __init__(self, reservation_id, gate):
        self.reservation_id = reservation_id
        self.bags = []  
        self.ready = False  
        self.gate = None

    def add_bag(self, bag):
        
        if isinstance(bag, Bag):
            self.bags.append(bag)
            bag.reservation = self
            self.check_ready_status()

    def check_ready_status(self):
        
        for i in range(0, len(self.bags)):
            if(self.bags[i].ready == False):
                return
            
        self.ready = True 
        
    def list_bags(self):
        if not self.bags:
            return "No bags added yet."
        return "\n".join(str(bag) for bag in self.bags)

    def set_gate(self, gate)
        self.gate = gate
        

    def __str__(self):
        return (f"Reservation(ID: {self.reservation_id}, Customer: {self.customer_name}, "
                f"Bags: {len(self.bags)}, Ready: {self.ready})")
```

```python
class PickupGate:
    
    def __init__(self, name, capacity):
        self.name = name 
        self.capacity = capacity
        self.res = None
        self.load = 0
        
        
    def add_reservation(self, reservation):
        if self.res is None:  
            self.res = reservation
            return True
        else
            return False


    def add_luggage(self, reservation, bag):
        if bag.reservation == reservation:
            self.load += 1
            return True

        print(f"Gate {self.name} has no space for a new reservation!")
        return False

    def remove_luggage(self):
        self.res = None
        self.load = 0



gate7A = PickupGate(name="7A", capacity=7)
gate7B = PickupGate(name="7B", capacity=7)
gate7C = PickupGate(name="7C", capacity=7)
gate6A = PickupGate(name="6A", capacity=6)
gate6B = PickupGate(name="6B", capacity=6)
gate6C = PickupGate(name="6C", capacity=6)
gate5A = PickupGate(name="5A", capacity=5)
gate5B = PickupGate(name="5B", capacity=5)
gate5C = PickupGate(name="5C", capacity=5)
gate4A = PickupGate(name="4A", capacity=4)
gate4B = PickupGate(name="4B", capacity=4)
gate4C = PickupGate(name="4C", capacity=4)
gate3A = PickupGate(name="3A", capacity=3)
gate3B = PickupGate(name="3B", capacity=3)
gate3C = PickupGate(name="3C", capacity=3)
gate2A = PickupGate(name="2A", capacity=2)
gate2B = PickupGate(name="2B", capacity=2)
gate2C = PickupGate(name="2C", capacity=2)



def find_gates(size, gates)
    row = gates[size-2]
    for gate in row
        if gate.res == None
            return gate
    return None


gates = [[gate2A, gate2B, gate2C], [gate3A, gate3B, gate3C], 
         [gate4A, gate4B, gate4C], [gate5A, gate5B, gate5C], 
         [gate6A, gate6B, gate6C], [gate7A, gate7B, gate7C]]
```
